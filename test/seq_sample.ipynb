```python
import random

# 生成时间序列
seq = pd.date_range('2019-10-30 00:00:00', '2019-10-30 02:30:00', freq="5min")

# 定义子序列长度
subseq_len = 25

# 计算可生成的子序列数量
num_subseq = len(seq) - subseq_len + 1
print(f"总共可以生成的子序列数: {num_subseq}")

# 划分为长度为25的滑动子序列
subseqs = [seq[i:i+subseq_len] for i in range(num_subseq)]

# 对subseqs进行shuffle
# random.shuffle(subseqs)

# 示例：打印第一个和最后一个子序列
print("第一个子序列：", subseqs[0])
print("最后一个子序列：", subseqs[-1])

subseqs = [seq[i:i+subseq_len][0] for i in range(num_subseq)]
subseqs
```
